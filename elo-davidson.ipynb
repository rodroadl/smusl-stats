```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from datetime import datetime
```

```python
KAPPA = 2
```

```python
def calculate_delta(r, kappa=2, s=400):
    return 10**(r/s) / (10**(-r/s) + kappa + 10**(r/s))

def round_to_nearest_even(num):
    ceil_num = math.ceil(num)
    floor_num = math.floor(num)
    return ceil_num if ceil_num - num < 0.5 else floor_num
```

```python
X = range(-500, 501)
win = []
draw = []
lose = [] 
draw_and_win = []

for x in X:
    # w = round_to_nearest_even(calculate_delta(x) * 100)
    # l = round_to_nearest_even(calculate_delta(-x) * 100)
    w = (calculate_delta(x) * 100)
    l = (calculate_delta(-x) * 100)
    dnw = 100 - l
    d = dnw - w

    win.append(w)
    lose.append(l)
    draw_and_win.append(dnw)
    draw.append(d)

plt.plot(X, win)
# plt.plot(X, lose)
plt.plot(X, draw_and_win)
# plt.plot(X, draw)

plt.show()
```

```python
X = list(range(-500, 501, 50))
win = []
draw = []
lose = [] 
draw_and_win = []

for x in X:
    # w = round_to_nearest_even(calculate_delta(x) * 100)
    # l = round_to_nearest_even(calculate_delta(-x) * 100)
    w = (calculate_delta(x) * 100)
    l = (calculate_delta(-x) * 100)
    dnw = 100 - l
    d = dnw - w

    win.append(w)
    lose.append(l)
    draw_and_win.append(dnw)
    draw.append(d)

bottom = np.zeros(21)
odds = {"Win": win ,"Draw":draw, "Lose": lose}
width = 40

fig, ax = plt.subplots(figsize=(15,10))

for result, odd in odds.items():
    p = ax.bar(X, odd, width, label=result, bottom=bottom)
    bottom += odd

    ax.bar_label(p, label_type='center', fmt='{:0.1f}')
    ax.set_xticks(X)

ax.legend()
plt.show()
```
